## first define speed function !

```python
def calculateDistance(x,y): 
    travel=0
    for i in range(len(y)-1):
        dist = math.sqrt((x[0+i] - x[1+i])**2 + (y[0+i] - y[1+i])**2)
        travel+=dist
        
    return travel  



from statistics import mean

def calculateSpeed(x,y,time): 
    travel=0
    speed=[] 
    for i in range(len(y)-1):
        dist = math.sqrt((x[0+i] - x[1+i])**2 + (y[0+i] - y[1+i])**2)/time[0+i]
        speed.append(dist)   
    return (mean(speed)) 
    
import numpy
Day79_fs1[0][1]-Day79_fs1[0][0]
time = numpy.arange(0.01, len(Day79_fs1[0]), 0.01)

    
# average speed per minute in meters
print (60 * 100* calculateSpeed(Day79_fs1[1],Day79_fs1[3],time))
```
