## Now Graph beacons visible vs. invisible over particular session. 

```python
def beacons_over_session (beacons_vis,beacons_invis,list_of_number_of_beacons,session) :
    """this function takes lists of days for each animal and plots a distance covered over time. """
    
    
    x = np.arange(len(list_of_number_of_beacons))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots(dpi= 500)
    FS1 = ax.bar(x - width/2, beacons_vis, width, label='visible '+ str(int(np.sum(beacons_vis)))+' s')
    FS2 = ax.bar(x + width/2, beacons_invis, width, fill=False, label='invisible '+ str(int(np.sum(beacons_invis)))+' s')

   
    ax.set_ylabel('seconds')
    ax.set_xlabel('beacon#')
    ax.set_title('Time to get beacon over session')
    ax.set_xticks(x)
    ax.set_xticklabels(list_of_number_of_beacons)
    ax.legend(loc='upper left',prop={'size': 6})


    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(int(height)),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 0),  # 3 points vertical offset - set to 0 
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(FS1)
    autolabel(FS2)

    fig.tight_layout()
    plt.savefig(figures+'beacons_over_session_'+ session +'.png', dpi = 1000)
    plt.show()
    
```

## Now graphing function

```python
beacons = [beacon_Day86_fs1,beacon_Day87_fs1,beacon_Day88_fs1,beacon_Day89_fs1,beacon_Day90_fs1,]
Day_number_list =('86','87','88','89','90')

def Total_beacons_over_sessions (list_of_beacon_days,list_of_number_of_days,animal_ID) :
    """this function takes lists of days for each animal and plots a distance covered over time. """
    

    
    visible=[]
    invisible = []
    

    for beacon in list_of_beacon_days:
        diff=np.diff(beacon[0])
        visible.append(sum(diff[1::2]))
        invisible.append (sum(diff[0::2]))
    
    x = np.arange(len(list_of_number_of_days))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots(dpi= 500)
    FS1 = ax.bar(x - width/2, visible, width, label='visible '+ str(int(np.sum(visible)/60))+' min')
    FS2 = ax.bar(x + width/2, invisible, width, fill=False, label='invisible '+ str(int(np.sum(invisible)/60))+' min')

   
    ax.set_ylabel('seconds looking for beacon')
    ax.set_xlabel('session#')
    ax.set_title('Time to get beacon over session '+ animal_ID)
    ax.set_xticks(x)
    ax.set_xticklabels(list_of_number_of_days)
    ax.legend(loc='upper left',prop={'size': 5})


    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(int(height)),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 0),  # 3 points vertical offset - set to 0 
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(FS1)
    autolabel(FS2)

    fig.tight_layout()
    plt.savefig(figures+'Total_beacons_over_session_'+ animal_ID +'_'+Day_number_list[0]+'-'+Day_number_list[-1]+'.png', dpi = 1000)
    plt.show()
```

```python
beacons = [beacon_Day86_fs1,beacon_Day87_fs1,beacon_Day88_fs1,beacon_Day89_fs1,beacon_Day90_fs1,beacon_Day91_fs1,beacon_Day92_fs1,beacon_Day93_fs1]
Day_number_list =('86','87','88','89','90','91','92','93')
Animal  = 'FS1'    
Total_beacons_over_sessions(beacons,Day_number_list,Animal)
```
