```python
def beacons_over_session (list_of_fs1_days,list_of_fs2_days,list_of_number_of_days,session) :
    """this function takes lists of days for each animal and plots a distance covered over time. """
    
    
    x = np.arange(len(list_of_number_of_days))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots(dpi= 500)
    FS1 = ax.bar(x - width/2, list_of_fs1_days, width, label='visible '+ str(int(np.sum(list_of_fs1_days)))+' s')
    FS2 = ax.bar(x + width/2, list_of_fs2_days, width, fill=False, label='invisible '+ str(int(np.sum(list_of_fs2_days)))+' s')

   
    ax.set_ylabel('seconds')
    ax.set_xlabel('beacon#')
    ax.set_title('Time to get beacon over session')
    ax.set_xticks(x)
    ax.set_xticklabels(list_of_number_of_days)
    ax.legend(loc='upper left',prop={'size': 6})


    def autolabel(rects):
        """Attach a text label above each bar in *rects*, displaying its height."""
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(int(height)),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 0),  # 3 points vertical offset - set to 0 
                        textcoords="offset points",
                        ha='center', va='bottom')


    autolabel(FS1)
    autolabel(FS2)

    fig.tight_layout()
    plt.savefig(figures+'beacons_over_session_'+ session +'.png', dpi = 1000)
    plt.show()
    
```
